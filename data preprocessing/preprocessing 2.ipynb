```python
import ast
data['genres']=data['genres'].map(lambda x : ast.literal_eval(x))
data['cast']=data['cast'].map(lambda x : ast.literal_eval(x))
data['crew']=data['crew'].map(lambda x : ast.literal_eval(x))
```

```python
def genres_list(x):
    gen=[]
    for i in x:
        if i.get('name')=='Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen==[]:
        return np.NaN
    else:
        return (" ".join(gen))
```

```python
data['genres']=data['genres'].map(lambda x:genres_list(x))
```

```python
data.head(10)
```

```python
def actor_1 (x):
    casts=[]
    for i in x:
        casts.append(i.get('name'))
    if casts==[]:
        return np.NaN
    else:
        return (casts[0])
```

```python
def actor_2 (x):
    casts=[]
    for i in x:
        casts.append(i.get('name'))
    if casts==[] or len(casts)<=1:
        return np.NaN
    else:
        return (casts[1])
```

```python
def actor_3 (x):
    casts=[]
    for i in x:
        casts.append(i.get('name'))
    if casts==[] or len(casts)<=2:
        return np.NaN
    else:
        return (casts[2])
```

```python
data['actor_1_name']=data['cast'].map(lambda x: actor_1(x))
```

```python
data['actor_2_name']=data['cast'].map(lambda x: actor_2(x))
```

```python
data['actor_3_name']=data['cast'].map(lambda x: actor_3(x))
```

```python
data['crew'][1]
```

```python
def get_directors(x):
    direct=[]
    for i in x:
        if(i.get('job')=='Director'):
            direct.append(i.get('name'))
    if direct==[]:
        return np.NaN
    else:
        return(" ".join(direct))
```

```python
data['director_name']=data['crew'].map(lambda x: get_directors(x))
```
